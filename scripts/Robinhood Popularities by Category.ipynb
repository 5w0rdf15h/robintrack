```python
import pandas as pd

import os
from datetime import datetime

def show_plot(file_path):
    df = pd.read_csv(file_path, names=["timestamp", "popularity"])
    df['timestamp'] = df['timestamp'].map(lambda dt_string: datetime.strptime(dt_string, '%Y-%m-%d'))

    # The following is based off of https://stackoverflow.com/a/42372823/3833068
    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates

    # plot
    plt.plot(df['timestamp'], df['popularity'])
    name = file_path.replace('.csv', '').replace('/tmp/out/', '')
    plt.title(f'{name} Popularity over Time')
    plt.ylabel('Unique Robinhood Users Holding Shares')
    # beautify the x-labels
    fig = plt.gcf()
    fig.autofmt_xdate()
    fig.set_size_inches(9, 5)

    print(file_path)
    plt.show()
    
for file_path in os.listdir('/tmp/out'):
    show_plot(f"/tmp/out/{file_path}")
```
